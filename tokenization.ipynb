```python
from transformers import BertTokenizer
import torch

# Load the pre-trained BERT tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Sample text (you can use your own dataset here)
texts = [
    "Natural Language Processing with BERT is powerful.",
    "Word embeddings capture semantic similarity."
]

# Tokenize and embed the text
encodings = tokenizer.batch_encode_plus(
    texts,
    padding=True,
    truncation=True,
    return_tensors='pt',  # Return PyTorch tensors
    max_length=128  # Set a maximum length for tokenization
)

# Extract token_ids, attention_mask, and token_type_ids
token_ids = encodings['input_ids']
attention_mask = encodings['attention_mask']
token_type_ids = encodings['token_type_ids']

# Print the tokenized values and embeddings
print("Token IDs:", token_ids)
print("Attention Mask:", attention_mask)
print("Token Type IDs:", token_type_ids)

# You can feed these embeddings to the BERT model for further processing
```
