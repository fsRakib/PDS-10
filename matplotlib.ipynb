Exploratory Data Analysis and Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
iris
```

```python
df=pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df.sample(5)
```

```python
df["Species"] = iris.target
df['Species'].unique()
```

```python
df['Species']=df['Species'].map({0: 'Setosa', 1:'Versicolor', 2:'Virginica'  })
```

```python
#random 5 row slection
df.head(200)
```

Exploratory Data Analysis

```python
x= df['sepal length (cm)']
y=df['sepal width (cm)']
```

```python
x[:3], y[:3]
```

```python
#visualize the sepal size
plt.scatter(x,y,c='green')
plt.title("Sepal Length vs Width")
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
color_map=df['Species'].map({'Setosa': 'red', 'Virginica':'green', 'Versicolor': 'blue'})
```

```python
#visualize by sepal size
plt.scatter(x,y,c=color_map)
plt.title("Sepal Length vs Width")
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
X=df['petal length (cm)']
Y=df['petal width (cm)']
print(type(X), type(Y))
X, Y
```

```python
# visualize by petal size
plt.scatter(df['petal length (cm)'], df['petal width (cm)'], c=df['Species'].map({'Setosa': 'red', "Versicolor": 'green', 'Virginica': 'blue'}))
plt.title("Petal Length vs Width by Species")
plt.xlabel("Petal Length (cm)")

plt.ylabel("Petal Width (cm)")
plt.show()
```

```python
#Box plot by sepal length
df.boxplot(column=['sepal length (cm)', 'sepal width (cm)'], by='Species', figsize=(8,6))
plt.show()
```

```python
df.boxplot(column=['petal length (cm)', 'petal width (cm)'], by='Species', figsize=(8,6))
plt.show()
```

```python
#Histogram
df.hist(bins=10, figsize=(10,8))
plt.tight_layout()
plt.show()
```

```python
#Data Original
ages=df['Age']
fares=df['Fare']

#Data Cureted
ages_files=df_filed['Age']
fares_files=df_filed['Fare']
```

```python
#Bar plot original
plt.subplot(1,2,1)
plt.bar(ages,fares,color='skyblue')
plt.title('Age vs Fare Original')
plt.xlabel('Age')
plt.ylabel('Fare')

#Bar plot filled
plt.subplot(1,2,2)
plt.bar(ages_files, fares_files, color='green')
plt.title('Age vs Fare Filled')
plt.xlabel('Age')
plt.ylabel('Fare')
```
